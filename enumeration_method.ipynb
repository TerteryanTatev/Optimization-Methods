```python
import sympy as sp
import matplotlib.pyplot as plt
```

```python
def analytical_extrema(x, f):
    derivative = sp.diff(f, x)
    extrema = sp.solveset(derivative, x, domain=sp.S.Reals)
    return extrema
```

```python
def enumeration_method(x, f, a, b, eps):
    n = int((b - a) / eps)  # number of steps
    points = []

    for i in range(n + 1):
        xi = a + i * (b - a) / n
        yi = f.subs(x, xi)
        points.append((xi, yi))

    # find point with minimal function value
    minimum = min(points, key=lambda point: point[1])

    return points, minimum
```

```python
def plot_results(points):
    x_vals = [p[0] for p in points]
    y_vals = [p[1] for p in points]

    plt.figure(figsize=(7, 4))

    # plot line between points
    plt.plot(x_vals, y_vals, linestyle='-', linewidth=2)

    # highlight points
    plt.scatter(x_vals, y_vals, s=40, zorder=3, color='darkblue')

    plt.title("Enumeration Method")
    plt.xlabel("x")
    plt.ylabel("f(x)")

    plt.grid(True, linestyle='-')
    plt.tight_layout()
    plt.show()

```

```python
def main():
    """
    Main program execution.
    """

    x = sp.symbols('x')
    f = (1/4) * x**4 + x**2 - 8*x + 12

    # compute analytical extrema (for reference)
    extrema = analytical_extrema(x, f)
    print("Analytical extrema:", extrema)

    # interval and precision
    a, b = 0, 2
    eps = 0.5

    # apply enumeration method
    points, minimum = enumeration_method(x, f, a, b, eps)

    print("Sampled points:", points)
    print("Approximate minimum:", minimum)

    # visualize results
    plot_results(points)

```

```python
if __name__ == "__main__":
    main()
```
