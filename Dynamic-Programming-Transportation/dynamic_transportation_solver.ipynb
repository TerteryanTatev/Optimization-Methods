```python
import math
import pandas as pd
from tabulate import tabulate

a1, a2 = 40, 60
b = [18, 14, 8, 16, 14, 15, 15]

alpha = [
    [1.0, 1.2, 1.3, 1.4, 1.5, 1.4, 1.3],
    [2.1, 2.2, 2.3, 2.4, 2.5, 2.4, 2.3]
]

beta = [
    [4.1, 4.2, 4.3, 4.4, 5.2, 5.5, 5.6],
    [3.2, 4.1, 4.2, 4.3, 4.4, 4.5, 4.6]
]

dp = {}
choice = {}

def F(j, s1, s2):
    if j == len(b):
        return 0

    if (j, s1, s2) in dp:
        return dp[(j, s1, s2)]

    demand = b[j]
    best = math.inf
    best_x1, best_x2 = 0, 0

    for x1 in range(max(0, demand - s2), min(s1, demand) + 1):
        x2 = demand - x1

        cost = (
            alpha[0][j] * x1 + beta[0][j] * x1**2 +
            alpha[1][j] * x2 + beta[1][j] * x2**2
        )

        total = cost + F(j + 1, s1 - x1, s2 - x2)

        if total < best:
            best = total
            best_x1, best_x2 = x1, x2

    dp[(j, s1, s2)] = best
    choice[(j, s1, s2)] = (best_x1, best_x2)

    return best


result = F(0, a1, a2)
print("\n Minimum cost =", round(result, 2))


def reconstruct():
    s1, s2 = a1, a2
    rows = []

    for j in range(len(b)):
        x1, x2 = choice[(j, s1, s2)]
        rows.append([j+1, s1, s2, x1, x2])

        s1 -= x1
        s2 -= x2

    return pd.DataFrame(rows, columns=[
        "Step (j)", "s1 (before)", "s2 (before)", "x1", "x2"
    ])


opt_df = reconstruct()

print("\n Optimal solution table:\n")
print(tabulate(opt_df, headers='keys', tablefmt='fancy_grid', showindex=False))



dp_rows = []

for (j, s1, s2), val in dp.items():
    dp_rows.append([j, s1, s2, round(val, 2)])

dp_df = pd.DataFrame(dp_rows, columns=["j", "s1", "s2", "F(j,s1,s2)"])
dp_df = dp_df.sort_values(by=["j", "s1", "s2"])

for j_value in sorted(dp_df["j"].unique()):
    print(f"\n DP Table for j = {j_value}:\n")

    sub_df = dp_df[dp_df["j"] == j_value]

    print(tabulate(sub_df, headers='keys', tablefmt='fancy_grid', showindex=False))
```
