```python
import numpy as np
from tabulate import tabulate

W_values = np.array([12, 18, 24, 42])       # Պատահական առաջարկների արժեքները
P_values = np.array([0.15, 0.25, 0.3, 0.3]) # Համապատասխան հավանականությունները
N_periods = 8                               # Փուլերի (հորիզոնի) քանակը
final_rejection_value = 0                  # 8-րդ փուլում մերժելու արժեքը

E_v = np.zeros(N_periods + 1)
E_v[N_periods] = final_rejection_value # E[v_9] = 0

history = []

for t in range(N_periods, 0, -1):
    R_t = E_v[t]

    v_t_values = np.maximum(W_values, R_t)

    E_v_current = np.sum(v_t_values * P_values)
    E_v[t-1] = E_v_current

    v_t_str = "(" + ", ".join([f"{val:.2f}" for val in v_t_values]) + ")"
    history.append({
        't': t,
        'R_t': R_t,
        'v_t': v_t_str,
        'E_next': R_t,
        'E_curr': E_v_current
    })

history = history[::-1]

print("\n" + "="*90)
print(f" ԴԻՆԱՄԻԿ ԾՐԱԳՐԱՎՈՐՄԱՆ ՀԱՇՎԱՐԿ (Periods = {N_periods})")
print("="*90)

table_data = []
for h in history:
    table_data.append([
        h['t'],
        f"{h['R_t']:.4f}",
        h['v_t'],
        f"{h['E_next']:.4f}",
        f"{h['E_curr']:.4f}"
    ])

headers = ["Փուլ (t)", "Ամրագրման արժեք (R_t)", "v_t(12, 18, 24, 42) արժեքներ", "Հաջորդի սպասվող E[v_{t+1}]", "Ընթացիկ սպասվող E[v_t]"]
print(tabulate(table_data, headers=headers, tablefmt="fancy_grid"))

print(f"\n Խնդրի սկզբում օպտիմալ սպասվող արժեքը (E[v_1]) = {E_v[0]:.4f}\n")



```
